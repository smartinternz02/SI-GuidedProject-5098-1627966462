```python
#Importing the libraries 
import numpy as np
import pandas as pd
import sklearn.metrics as mse
# Importing libraries to handle categorical data 
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
# Importing libraries to split into train and test sets 
from sklearn.model_selection import train_test_split
#Importing libraries to handle feature scaling 
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv("50_Startups.csv")
df.head()
```

```python
#checking for missing data if there is any and how many 
df.isnull().sum()
```

```python
df.info()
```

```python
#since we got some missing values we need to handle them so we need to do following
# Replace missing values  with the mean (because data is numerical)
df['Administration'].fillna(df['Administration'].mean(), inplace = True)
df['Marketing Spend'].fillna(df['Marketing Spend'].mean(), inplace = True)
```

```python
# so as we see now the dataset has zero missing value so we can proceed
df.isnull().sum()
```

```python
#Splitting data into inputs as x and outputs as y
x = df.iloc[:, 0:4]
x.head()
```

```python
y = df.iloc[:, -1:].values
y
```

```python
#Handling categorical data and convert them into binary
ct = ColumnTransformer([('one', OneHotEncoder(), [3])], remainder = 'passthrough')
x = ct.fit_transform(x)
```

```python
df["State"].unique()
```

```python
x
```

```python
# Splitting data to train sets and test sets
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2, random_state = 0)
```

```python
print('Inputs shape ', x.shape)
print('Outputs shape ', y.shape)
print('x_train shape ', x_train.shape)
print('x_test shape ', x_test.shape)
print('y_train shape ', y_train.shape)
print('y_test shape', y_test.shape)
```

```python
#feature scaling 
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.fit_transform(x_test)
```

```python
# DL libraries
from tensorflow.keras.models import Sequential 
from tensorflow.keras.layers import Dense 
```

```python
ob=Sequential()
ob.add(Dense(units=6,kernel_initializer= "random_uniform",activation= "relu"))# Input layer
ob.add(Dense(units=12,kernel_initializer= "random_uniform",activation= "relu"))# Hidden layer 1
ob.add(Dense(units=12,kernel_initializer= "random_uniform",activation= "relu"))# Hidden layer 2 
ob.add(Dense(units=12,kernel_initializer= "random_uniform",activation= "relu"))# Hidden layer 3 
ob.add(Dense(units=12,kernel_initializer= "random_uniform",activation= "relu"))# Hidden layer 4 
ob.add(Dense(units=12,kernel_initializer= "random_uniform",activation= "relu"))# Hidden layer 5
ob.add(Dense(units=1,kernel_initializer= "random_uniform",activation= "relu"))#  Output layer 
ob.compile(optimizer="rmsprop",loss="mean_squared_error")# Compiling model 
```

```python
ob.fit(x_train, y_train, epochs = 1000)# Training the model 
```

```python
ypred = ob.predict(x_test)
ypred
```
